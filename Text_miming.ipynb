```python
from textblob import TextBlob
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer  # Importez WordNetLemmatizer
import pandas as pd

# Votre fonction de prétraitement des tweets
def preprocess_text(tweet):
    blob = TextBlob(tweet)
    tokens = blob.words  # Utilisation de TextBlob pour la tokenisation
    stop_words = set(stopwords.words('english'))
    processed_tokens = [token.lower() for token in tokens if token.lower() not in stop_words and token.lower() not in string.punctuation]
    lemmatized_tokens = [WordNetLemmatizer().lemmatize(word, pos='v') for word in processed_tokens]  # Lemmatisation avec NLTK
    return ' '.join(lemmatized_tokens)

# Chargement et prétraitement de votre jeu de données
data = pd.read_csv('C:\\Users\\DELL\\OneDrive\\Documents\\LICENCE_2_DIT_BIG_DATA\\Dossier_PROJET_FIN.csv')
data['processed_text'] = data['text'].apply(preprocess_text)

# Sauvegarde du DataFrame prétraité si nécessaire
data.to_csv('C:\\Users\\DELL\\OneDrive\\Documents\\LICENCE_2_DIT_BIG_DATA\\Dossier_PROJET_FIN.csv', index=False)
```

Analyse des sentiments
Pour attribuer une polarité (positive, négative, neutre) à chaque tweet


```python
from textblob import TextBlob

# Fonction pour déterminer la polarité d'un texte avec TextBlob
def get_sentiment(text):
    analysis = TextBlob(text)
    # Retourne la polarité comme 'positive', 'negative', ou 'neutral'
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'

# Appliquer la fonction sur chaque tweet et créer une nouvelle colonne 'sentiment' dans le DataFrame
df_preprocessed['sentiment'] = df_preprocessed['processed_text'].apply(get_sentiment)

# Afficher les premières lignes pour vérification
print(df_preprocessed[['processed_text', 'sentiment']].head())
```
